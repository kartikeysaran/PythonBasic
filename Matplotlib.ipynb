```python
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np

x= np.arange(-3, 3, 0.001)

plt.plot(x, norm.pdf(x))
plt.show()
```

```python
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.0, .5))
plt.savefig("c:\\Users\\lenovo\\Desktop\\Plot.png", format="png")
```

```python
axes = plt.axes()
axes.set_xlim([-5, 5])
axes.set_ylim([0, 1.0])
axes.set_xticks(np.arange(-5,5,1))
axes.set_yticks(np.arange(0,1,0.1))
axes.grid()
plt.xlabel('Greebles')
plt.ylabel('Probability')
plt.plot(x, norm.pdf(x),'b-')
plt.plot(x, norm.pdf(x, 1,0.5), 'r:')
plt.legend(['Sneetches', 'Gacks'], loc=5)
plt.show()
```

```python
plt.rcdefaults()

values = [12, 55, 4, 32, 14]
colors = ['r', 'g', 'b', 'c', 'm']
explode = [0.1, 0, 0, 0, 0]
labels = ['India', 'United States', 'Russia', 'China', 'Europe']
plt.pie(values, colors=colors, labels= labels, explode = explode)
plt.title('Student Location')
plt.show()
```
